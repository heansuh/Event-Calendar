```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd

import locale
from datetime import datetime, timedelta
import time
```

### Dev

```python
driver = webdriver.Firefox()
```

```python
driver.get("https://rausgegangen.de/hamburg/kategorie/konzerte-und-musik/")
#driver.get("https://rausgegangen.de/en/hamburg/category/concerts-and-music/") #TODO later navigate to other cities & categories
```

```python
try:
    # Wait until the cookie rejection button is present and clickable
    wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds
    reject_button = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'iubenda-cs-reject-btn')))

    # Click the reject button
    reject_button.click()
    print("Cookie rejection button clicked successfully.")

except Exception as e:
    # Handle exceptions if the button is not found or not clickable
    print(f"An error occurred: {e}")
```

```python
try:
    # Wait until the button is present and clickable
    wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds
    submit_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//form[@action="/i18n/setlang/"]//button[@type="submit"]')))

    # Click the button
    submit_button.click()
    print("Form submit button clicked successfully.")

except Exception as e:
    # Handle exceptions if the button is not found or not clickable
    print(f"An error occurred: {e}")
```

```python
try:
    # Wait until the elements with class 'tile-medium lg:mb-2' are present
    wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds
    tiles = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'tile-medium')))

    # Loop through each tile and extract the required information
    for tile in tiles:
        # Extract the source (href)
        source = tile.find_element(By.TAG_NAME, 'a').get_attribute('href')
        
        # Extract the date and time
        date_time_element = tile.find_element(By.CLASS_NAME, 'text-sm')
        date_time_text = date_time_element.text.split('|')
        date = date_time_text[0].strip() if len(date_time_text) > 0 else ''
        time = date_time_text[1].strip() if len(date_time_text) > 1 else ''

        # Extract the event name
        event_name = tile.find_element(By.CLASS_NAME, 'text-truncate--2').text

        # Extract the location
        location = tile.find_element(By.CLASS_NAME, 'opacity-70').text

        # Extract the price
        price = tile.find_element(By.CLASS_NAME, 'text-primary').text

        # Print the extracted information
        print(f"Source: {source}")
        print(f"Date: {date}")
        print(f"Time: {time}")
        print(f"Event Name: {event_name}")
        print(f"Location: {location}")
        print(f"Price: {price}")
        print("-" * 40)

except Exception as e:
    # Handle exceptions if any element is not found or any other error occurs
    print(f"An error occurred: {e}")
```

```python
try:
    # Wait until the elements with class 'tile-medium lg:mb-2' are present
    wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds
    tiles = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'tile-medium')))

    # List to store the extracted information
    data = []

    # Loop through each tile and extract the required information
    for tile in tiles:
        # Extract the source (href)
        source = tile.find_element(By.TAG_NAME, 'a').get_attribute('href')
        
        # Extract the date and time
        date_time_element = tile.find_element(By.CLASS_NAME, 'text-sm')
        date_time_text = date_time_element.text.split('|')
        date = date_time_text[0].strip() if len(date_time_text) > 0 else ''
        time = date_time_text[1].strip() if len(date_time_text) > 1 else ''

        # Extract the event name
        event_name = tile.find_element(By.CLASS_NAME, 'text-truncate--2').text

        # Extract the location
        location = tile.find_element(By.CLASS_NAME, 'opacity-70').text

        # Extract the price
        price = tile.find_element(By.CLASS_NAME, 'text-primary').text

        # Append the extracted information to the list
        data.append({
            "Source": source,
            "Date": date,
            "Time": time,
            "Event Name": event_name,
            "Location": location,
            "Price": price
        })

except Exception as e:
    # Handle exceptions if any element is not found or any other error occurs
    print(f"An error occurred: {e}")

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)
df = df[['Event Name', 'Date', 'Time', 'Location', 'Price', 'Source']]
df
```
