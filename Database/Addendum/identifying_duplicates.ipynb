```python
# Create df of two csv files
df1 = pd.read_csv('./sample_csvs/kiel_sailing_city_de.csv')
df2 = pd.read_csv('./sample_csvs/Kiel_wasgeht.csv')
```

```python
# Renaming columns in df1 to align with df2
df1 = df1.rename(columns={
    'Title': 'Subject',
    'Date': 'Start_date',
    'Time': 'Start_time',
    'Categories': 'Description'
})

# Display the updated DataFrame structure
df1.head()
```

```python
# Split Start_time into Start_time and End_time
df1[['Start_time', 'End_time']] = df1['Start_time'].str.extract(r'(\d{2}:\d{2})\s*-\s*(\d{2}:\d{2})')

# Display the updated DataFrame structure
df1.head()
```

```python
# Standardize Start_date to ISO 8601 format
df1['Start_date'] = pd.to_datetime(df1['Start_date'], format='%d.%m.%Y').dt.strftime('%Y-%m-%d')

# Display the updated DataFrame structure
df1.head()
```

```python
# Convert Description (Categories) from list to comma-separated string
df1['Description'] = df1['Description'].apply(lambda x: ', '.join(x) if isinstance(x, list) else str(x))

# Display the updated DataFrame structure
df1.head()
```

```python
# Ensure all required columns are in df1
required_columns = ['Subject', 'Start_date', 'Start_time', 'End_time', 'Location', 'Description']

# Add any missing columns with default values (NaN for now)
for col in required_columns:
    if col not in df1.columns:
        df1[col] = pd.NA

# Ensure column order matches df2
df1 = df1[required_columns]

# Display the updated DataFrame structure
df1.head()
```

```python
# Concatenate df1 and df2
combined_df = pd.concat([df1, df2], ignore_index=True)

# Display the structure of the combined DataFrame
combined_df.head()
```

```python
combined_df.to_csv('./sample_csvs/combined.csv', index=False)
```

#### **Key Characteristics Relevant to Deduplication**
1. **Duplicates**:
   - Likely occur due to events happening daily over a period (e.g., recurring events).
   - Identifiable by shared `Subject` and `Location`, with overlapping or consecutive `Start_date` values.

2. **Time Columns**:
   - `Start_time` is mostly present, but `End_time` is often missing.
   - Events without `End_time` might still overlap with others based on `Start_date` and `Location`.

3. **Locations**:
   - Varied formats and naming conventions (e.g., "CITTI-PARK" vs. "CITTI Park Kiel").
   - Requires normalization for robust duplicate detection.

4. **Descriptions**:
   - Contains a mix of structured and unstructured content:
     - Lists (e.g., categories like `['Ausstellung', 'Kultur']`).
     - Plain text with potential tags (e.g., "concert").
   - Standardizing or extracting structured information may aid deduplication.

```python
import re

def normalize_text(text, is_location=True, remove_suffix=None):
    """
    Normalize text strings for either locations or subjects.
    
    Args:
        text (str): The original string.
        is_location (bool): Whether the text is a location (default True).
        remove_suffix (str): A specific suffix to remove (e.g., ', kiel').
    
    Returns:
        str: The normalized string.
    """
    # Lowercase the string
    text = text.lower()
    # Replace punctuation with spaces, except "&"
    text = re.sub(r"[^\w\s&]", " ", text)
    # Replace multiple spaces with a single space
    text = re.sub(r"\s+", " ", text).strip()
    
    # Remove specific suffix if requested
    if remove_suffix and text.endswith(remove_suffix):
        text = text[: -len(remove_suffix)].strip()
    
    return text
```

```python
def extract_tokens(text):
    """
    Extract significant tokens (key words) for deduplication purposes.
    
    Args:
        text (str): The normalized string.
    
    Returns:
        str: Tokenized string with key words.
    """
    # Split into tokens
    tokens = text.split()
    # Retain only unique words (ignoring order) for deduplication purposes
    return " ".join(sorted(set(tokens)))
```

```python
# Example usage for Subjects
combined_df['Subject_normalized'] = combined_df['Subject'].apply(
    lambda sub: normalize_text(sub, is_location=False)
)
combined_df['Subject_tokens'] = combined_df['Subject_normalized'].apply(extract_tokens)

# Example usage for Locations
combined_df['Location_normalized'] = combined_df['Location'].apply(
    lambda loc: normalize_text(loc, is_location=True, remove_suffix=", kiel")
)
combined_df['Location_tokens'] = combined_df['Location_normalized'].apply(extract_tokens)

# Display results
combined_df[['Subject', 'Subject_normalized', 'Subject_tokens', 'Location', 'Location_normalized', 'Location_tokens']].head()
```

```python
from Levenshtein import distance as levenshtein_distance

def tokens_match(token1, token2, max_distance=1):
    """
    Compare two tokens using Levenshtein distance.
    
    Args:
        token1 (str): First token.
        token2 (str): Second token.
        max_distance (int): Maximum allowed edit distance for a match.
    
    Returns:
        bool: True if tokens match, False otherwise.
    """
    return levenshtein_distance(token1, token2) <= max_distance
```

```python
def jaccard_similarity(set1, set2):
    """
    Compute Jaccard similarity between two sets.
    
    Args:                               
        set1 (set): First set of tokens.
        set2 (set): Second set of tokens.
    
    Returns:                                                        
        float: Jaccard similarity score.
    """
    intersection = len(set1 & set2)
    union = len(set1 | set2)
    return intersection / union if union > 0 else 0
```

```python
def compare_events(tokens_a, tokens_b, token_threshold=1, jaccard_threshold=0.8):
    """
    Compare two sets of tokens for event similarity.
    
    Args:
        tokens_a (str): Tokenized string for event A.
        tokens_b (str): Tokenized string for event B.
        token_threshold (int): Maximum allowed Levenshtein distance for token matching.
        jaccard_threshold (float): Minimum Jaccard similarity for a match.
    
    Returns:
        bool: True if events are similar, False otherwise.
    """
    # Split into individual tokens
    set_a = set(tokens_a.split())
    set_b = set(tokens_b.split())
    
    # Find token matches based on Levenshtein distance
    matched_tokens = set()
    for token_a in set_a:
        for token_b in set_b:
            if tokens_match(token_a, token_b, max_distance=token_threshold):
                matched_tokens.add(token_a)
                matched_tokens.add(token_b)
    
    # Compute Jaccard similarity on matched tokens
    jaccard_score = jaccard_similarity(set_a, set_b)
    
    # Return True if Jaccard similarity exceeds the threshold
    return jaccard_score >= jaccard_threshold
```

```python
from itertools import combinations

# Function to deduplicate the entire DataFrame
def deduplicate_events(df, token_threshold=1, jaccard_threshold=0.8):
    """
    Deduplicate events based on Subject and Location tokens.

    Args:
        df (pd.DataFrame): The DataFrame containing event data.
        token_threshold (int): Max Levenshtein distance for token matching.
        jaccard_threshold (float): Min Jaccard similarity for a match.

    Returns:
        pd.DataFrame: Original DataFrame with an added 'Duplicate_Group' column.
    """
    # Initialize a column for duplicate group IDs
    df['Duplicate_Group'] = -1  # Default: no group

    # Get all pairs of event indices for comparison
    event_pairs = combinations(df.index, 2)

    # Dictionary to track duplicate group assignments
    group_dict = {}
    group_id = 0

    for idx1, idx2 in event_pairs:
        # Extract tokens for the pair of events
        subject1, subject2 = df.loc[idx1, 'Subject_tokens'], df.loc[idx2, 'Subject_tokens']
        location1, location2 = df.loc[idx1, 'Location_tokens'], df.loc[idx2, 'Location_tokens']

        # Compare subject and location tokens
        subject_similar = compare_events(subject1, subject2, token_threshold, jaccard_threshold)
        location_similar = compare_events(location1, location2, token_threshold, jaccard_threshold)

        # If both subject and location are similar, group them
        if subject_similar and location_similar:
            # Check if either event is already in a group
            if idx1 in group_dict:
                group_dict[idx2] = group_dict[idx1]
            elif idx2 in group_dict:
                group_dict[idx1] = group_dict[idx2]
            else:
                # Assign a new group ID
                group_dict[idx1] = group_id
                group_dict[idx2] = group_id
                group_id += 1

    # Assign group IDs back to the DataFrame
    for idx, group in group_dict.items():
        df.at[idx, 'Duplicate_Group'] = group

    return df

# Apply deduplication to the full DataFrame
deduplicated_df = deduplicate_events(combined_df)

# Display results with duplicate groups
deduplicated_df[['Subject', 'Location', 'Duplicate_Group']].sort_values('Duplicate_Group')
```
